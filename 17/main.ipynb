```python
import pandas as pd
```

```python
df = pd.read_csv("AirQualityUCI.csv", delimiter=";", decimal=",")
df.head()
```

```python
df = df.drop(["Unnamed: 15", "Unnamed: 16"], axis=1)
df = df.dropna()
df.head()
```

```python
df
```

```python
df_adj = df.copy()
df_adj["Datetime"] = pd.to_datetime(df["Date"] + ' ' + df["Time"], format="%d/%m/%Y %H.%M.%S")
df_adj = df_adj.drop(["Date", "Time"], axis=1)
df_adj.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error
```

```python
df_adj.set_index('Datetime', inplace=True)
```

```python
df_adj[df_adj['C6H6(GT)'] < 0] = -1 
```

```python
air_quality_series = df_adj['C6H6(GT)']
```

```python
train_size = int(len(air_quality_series) * 0.8)
train, test = air_quality_series[:train_size], air_quality_series[train_size:]
```

```python
order = (1, 1, 1)  # You may need to tune these parameters based on your data
model = ARIMA(train, order=order)
fit_model = model.fit()
```

```python
predictions = fit_model.predict(start=len(train), end=len(train) + len(test) - 1, typ='levels')
```

```python
mae = mean_absolute_error(test, predictions)
print(f'Mean Absolute Error (MAE): {mae}')
```

```python
# Visualize the results
plt.plot(test.index, test, label='Actual', color='blue')
plt.plot(test.index, predictions, label='Predicted', color='red')
plt.title('Air Quality Forecasting with ARIMA')
plt.xlabel('Date')
plt.ylabel('Air Quality')
plt.legend()
plt.show()
```
