```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_train = pd.read_csv("dataset/train.csv")
df_test_X = pd.read_csv("dataset/test.csv")
df_test_y = pd.read_csv("dataset/gender_submission.csv")
```

```python
df_train.columns, df_test_X.columns, df_test_y.columns
```

Concatenate all dfs to dataset

```python
df_test = df_test_X.merge(df_test_y, how="inner", on="PassengerId")
df = df_train.merge(df_test, "outer")
df.head()
```

```python
df_enc = df.copy()
```

encode sex as binary / ordinal

```python
df_enc["sex_encoded"] = df["Sex"].astype("category").cat.codes
```

```python
df_enc.head()
```

encode embarked as one hot

```python
dummies = pd.get_dummies(df["Embarked"], prefix="Embarked")
dummies
```

```python
df_enc = pd.concat([df_enc, dummies], axis=1)
df_enc.head()
```
