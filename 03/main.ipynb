```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("iris.data", names=["sepal_length", "sepal_width", "petal_length", "petal_width", "class"])
df.head()
```

```python
df.describe()
```

```python
df["class"].value_counts()
```

```python
sns.pairplot(df, hue="class")
plt.show()
```

```python
df_melted = pd.melt(df, id_vars='class', var_name='measurement')
df_melted.head()
```

```python
plt.figure(figsize=(16, 10))
sns.catplot(x='measurement', y='value', hue='class', kind='box', data=df_melted, height=6, aspect=1.5)
plt.title('Boxplots of Sepal and Petal Measurements by Class')

plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

axes = axes.flatten()

class_order = df_melted['class'].unique()

for i, measurement in enumerate(['sepal_length', 'sepal_width', 'petal_length', 'petal_width']):
    sns.boxplot(x='class', y='value', data=df_melted[df_melted['measurement'] == measurement],
                ax=axes[i], order=class_order)
    axes[i].set_title(f'{measurement}')

plt.tight_layout()

plt.show()
```
