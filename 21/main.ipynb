```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import StandardScaler
```

```python
fashion_mnist = fetch_openml('Fashion-MNIST', as_frame=False)
X = fashion_mnist.data
y = fashion_mnist.target.astype(int)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
n_components = 2  # Number of components to keep (you can adjust this)
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(10, 8))
for i in range(10):  # Plot a few examples of each class
    indices = np.where(y == i)[0]
    plt.scatter(X_pca[indices, 0], X_pca[indices, 1], label=str(i))

plt.title('Fashion MNIST Dataset - PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```
